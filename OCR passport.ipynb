```python
import os
import requests
import httplib2

import cv2
import pytesseract

from pprint import pprint
from datetime import datetime


class Ya_Passports: # Класс для картинок паспортов с Яндекс Диска
    
    def __init__(self):
        self.Url = 'https://cloud-api.yandex.net/v1/disk/public/resources?public_key=https%3A%2F%2Fdisk.yandex.ru%2Fd%2FkVdTaY8TuGvBkw'

    def get_url_passports_img(self): # Получаем фото паспортов с Я.Диска
        headers = {'Content-Type': 'application/json', 'Accept': 'application/json', 'Authorization': 'OAuth'}
        request_passports = requests.get(self.Url, headers=headers).json()['_embedded']['items']

        # Возвращаем кортеж ссылок на фото паспортов
        tuple_urls_imgs = ()
        for item_passport in request_passports:
            file_img = item_passport['file']
            mime_type_img = item_passport['mime_type']
            mime_type_img = mime_type_img[mime_type_img.find('/')+1:]
            tuple_urls_imgs += (file_img +'.'+ mime_type_img,)
            
        return (tuple_urls_imgs)


def print_FIO(name_file):
    img = cv2.imread(name_file)
    ret, img = cv2.threshold(img, 127, 255, 0)
    text_img = pytesseract.image_to_string(img, lang='rus', config=r'--oem 3 --psm 6').capitalize()#.split("\n")
    print(text_img)
#         print(text_img[10], text_img[12], text_img[14])
#         cv2.imshow('Result', img) 4-11 13 14
#         cv2.waitKey(0)
    
    
def rectangle_img(image_path): # Обрежем временный файл по нужному тексту
    # Найдём лицо
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor= 1.1, minNeighbors=5, minSize=(10, 10))
    sp = image.shape # размер картинки
    
    # Размер нужной области текста
    x = (faces[0][0]*3)
    y = (faces[0][1]-faces[0][3])
    w = (faces[0][0]*8)
    h = int(faces[0][1]+faces[0][3]/2)
#     cv2.rectangle(image, (x, y), (w, h), (255, 255, 0), 2)
    # Обновим временный файл
    cropped = image[y:h, x:w]
    cropped_img = cv2.imwrite(image_path, cropped)
#     cv2.imshow('Result', image)
#     cv2.waitKey(60)
    
    
def get_passports_img(url_img):
    # Получим картинку во временный файл
    h = httplib2.Http('.cache')
    response, content = h.request(url_img)
    name_file =str(datetime.now) + '.jpg'
    with open(name_file, 'wb') as file_tmp:
        file_tmp.write(content)
        file_tmp.close()
    return name_file
    
    
if __name__ == '__main__':
    tuple_urls_imgs = Ya_Passports().get_url_passports_img()
    for url_img in tuple_urls_imgs:    
        name_file = get_passports_img(url_img) # Получим картинку во временный файл
        rectangle_img(name_file) # Выделим нужную область для чтения текста
        print_FIO(name_file) # Прочитаем ФИО
#         os.remove(name_file) # Удаляем временный файл
        print('-'*10)
    cv2.destroyAllWindows()
#         break
```
